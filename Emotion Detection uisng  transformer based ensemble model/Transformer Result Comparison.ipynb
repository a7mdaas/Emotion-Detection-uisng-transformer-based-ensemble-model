```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset/messages_dev_features_ready_for_WS_2022.tsv", sep="\t")
df.emotion.value_counts().plot(kind='bar', title= "Emotions data distribution")
```

```python
actual_label = {
    "anger": 0,
    "disgust": 1,
    "fear": 2,
    "joy": 3,
    "neutral": 4,
    "sadness": 5,
    "surprise": 6
}
```

```python
actual_label = df.emotion.replace(actual_label)
```

```python
pred_label = np.load("val_BERTBase_0.npy")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(actual_label, pred_label)
```

# BERT BASE with AOUS

```python
print(classification_report(actual_label, pred_label))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title("BERTBase with AOUS")
sns.heatmap(cm,
            annot=True,
            xticklabels=["anger","disgust","fear","joy","neutral","sadness","surprise"],
            yticklabels=["anger","disgust","fear","joy","neutral","sadness","surprise"])
```
